```python
import requests,pandas as pd
```

```python
df3['Arrival_Date']=pd.to_datetime(df3['Arrival_Date'])
```

```python
lat,lon = 20.1491, 74.2326
start=df3['Arrival_Date'].min().strftime('%Y-%m-%d')
end=df3['Arrival_Date'].max().strftime('%Y-%m-%d')
```

```python
print(start)
```

```python
url=f"https://archive-api.open-meteo.com/v1/archive?latitude={lat}&longitude={lon}&start_date={start}&end_date={end}&daily=precipitation_sum&timezone=auto"
```

```python
r=requests.get(url).json()
```

```python
r
```

```python
rain=pd.DataFrame({'date':r['daily']['time'],'precip_mm':r['daily']['precipitation_sum']})
```

```python
rain['date']=pd.to_datetime(rain['date'])
```

```python
df3=df3.merge(rain.rename(columns={'date':'Arrival_Date'}),on='Arrival_Date',how='left')
```

```python
df3['precip_mm'].max()
```

```python
fest=pd.read_csv('festival_dates_lasalgaon_onion_prices.csv',parse_dates=['festival_dates'])
```

```python
fest.head()
```

```python
len(fest)
```

```python
fest['festival_dates'].duplicated().sum()
```

```python
fest['festival_dates']=pd.to_datetime(fest['festival_dates']).dt.normalize()
```

```python
df3['Arrival_Date'].dtype
```

```python
df3['Arrival_Date']=df3['Arrival_Date'].dt.normalize()
```

```python
fest_dates_set=set(fest['festival_dates'])
```

```python
fest_dates_set
```

```python
df3['Arrival_Date'].nunique()
```

```python
df3['Arrival_Date'].count()
```

```python
shifts=[-2,-1,0,1,2]
rows=[]

for shift in shifts:
  temp=fest.copy()
  temp['festival_shifted']=temp['festival_dates']+pd.Timedelta(days=shift)
  temp['days_diff']=shift
  rows.append(temp[['festival_dates','fest','festival_shifted','days_diff']])

festival_expanded=pd.concat(rows,ignore_index=True)
```

```python
festival_expanded
```

```python
df_fest=df3.merge(festival_expanded,left_on='Arrival_Date',right_on='festival_shifted',how='left')
```

```python
df_fest['festival_window']=df_fest['fest'].notna().astype(int)
```

```python
df_fest=df_fest.sort_values(['Arrival_Date','days_diff'])
df_fest=df_fest.groupby('Arrival_Date',as_index=False).first()
```

```python
df_final=df_fest[['Arrival_Date','festival_window','fest','festival_dates','days_diff','Modal_Price','log_ret','vol_7','vol_30','month','week','day_of_week','precip_mm']]
```

```python
df_final.info()
```

```python
df_final['festival_window'].value_counts()
```

```python
df_final.drop(columns=['fest','festival_dates','days_diff'],inplace=True,errors='ignore')
```

```python
df_final.info()
```

```python
df_final['vol_30']=df_final['vol_30'].bfill()
```

```python
df_final.info()
```

```python
df_final['rainfall_7d']=df_final['precip_mm'].rolling(7).sum().fillna(0) # Calculate the sum of precipitation over the last 7 days and fill any missing values with 0.
```

```python
df_final['days_since_last_arrival']=df_final['Arrival_Date'].diff().dt.days.fillna(1) # Calculate the number of days since the last arrival date, filling the first value with 1.
df_final['days_since_last_arrival']=df_final['days_since_last_arrival'].clip(upper=10) # Cap the 'days_since_last_arrival' at 10 to limit extreme values.
```

```python
df_final['is_price_jump']=(df_final['log_ret'].abs()>2*df_final['vol_30']).astype(int) # Create a binary flag indicating if the absolute logarithmic return is more than twice the 30-day volatility.
```
